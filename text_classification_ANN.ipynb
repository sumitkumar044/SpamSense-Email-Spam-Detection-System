```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.neural_network import MLPClassifier
from sklearn.preprocessing import LabelEncoder
from matplotlib import pyplot as plt
import seaborn as sns
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,precision_score,recall_score,f1_score
```

```python
file_path = '/content/drive/MyDrive/DataScience/emailspam_detect.csv'
data=pd.read_csv(file_path)
```

```python
print(data.head())
```

```python
x=data['Emails']
y=data['Pridiction']
```

```python
print(x.shape,y.shape)
```

```python
#Encoding  0->nonspam , 1-> spam
le=LabelEncoder()
# y=le.fit_transform(y)
data['Pridiction'] = le.fit_transform(data['Pridiction'])
```

```python
print(data.head())
```

```python
count=data["Pridiction"].value_counts()
print(count)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
print("Training size",x_train.shape)
print("testing size",x_test.shape)
```

```python
vectorizer=CountVectorizer()
x_train_vectors=vectorizer.fit_transform(x_train)
x_test_vectors=vectorizer.transform(x_test)
```

```python
print(x_train_vectors.shape)
print(x_test_vectors.shape)
```

```python
vocabulary_from_vectorizer=vectorizer.vocabulary_
print(len(vocabulary_from_vectorizer))
```

```python
list(vocabulary_from_vectorizer.items())[:10]
```

```python
print(x_train_vectors.shape)
```

```python
new_model=MLPClassifier(hidden_layer_sizes=(100,100),
                        max_iter=1000,
                        activation='tanh',
                        solver='sgd',
                        learning_rate_init=0.001,
                        learning_rate='constant',
                        random_state=42,)
```

```python
new_model.fit(x_train_vectors,y_train)
```

```python
y_pred=new_model.predict(x_test_vectors)

accuracy=accuracy_score(y_test,y_pred)
print("Accuracy:",accuracy)
```

```python
confu_mat=confusion_matrix(y_test,y_pred)
class_report=classification_report(y_test,y_pred)
precision=precision_score(y_test,y_pred,average='weighted')
recall=recall_score(y_test,y_pred,average='weighted')
f1=f1_score(y_test,y_pred,average='weighted')
```

```python
print('confusion_matrix:',confu_mat)
print('classification_report:',class_report)
print('precision score:',precision)
print('recall score:',recall)
print('f1_score:',f1)
```
